Importing the different libraries


```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split  # without using any manual effort it will split the test cases
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

collection of data and its processing

```python
# loading dataset to a pandas data framework
sonar_data=pd.read_csv('/content/Copy of sonar data.csv',header=None) # hamari file me header (column ka name )nhi hai
```

```python
sonar_data.head()  # it will diplay first 5 rows of our data points
```

```python
sonar_data.shape
```

```python
sonar_data.describe() # it gives statistical measure of the data
```

```python
sonar_data[60].value_counts() # it will count how many data points for the rock and mine
```

```python
sonar_data.groupby(60).mean() # we are groupind data on the basis of rock and mine and finding mean for those groups
```

```python
#separating the data and labels as it is supervised learning algorithm
x=sonar_data.drop(columns=60,axis=1)  # we are just droping 60 th column as it is our lable
y=sonar_data[60]
```

```python
print(x)
print(y)
```

training and testing data

```python
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.1,stratify=y,random_state=1) # test_size =0.1 means 10 percentage of data is test data , stratify means dono rock and mine baraber me lena hai
```

```python
print(x.shape,x_train.shape, x_test.shape)
```

trainig the model---> logistic regression model

```python
model1=LogisticRegression()
```

```python
# training the data with the training set
model1.fit(x_train, y_train)
```

model evaluating

```python
# finding the accuracy on the training the data
x_training_prediction=model1.predict(x_train)
data_accuracy1=accuracy_score(x_training_prediction,y_train)
```

```python
print("accuracy : ",data_accuracy1 )
```

finding accuracy with test cases

```python
x_test_prediction=model1.predict(x_test)
data_accuracy2=accuracy_score(x_test_prediction,y_test)
```

```python
print("accuracy : ",data_accuracy2 )
```
